```python
from datasets import load_dataset
dataset_name = 'star-graph-deg-5-path-5-nodes-300'
dataset = load_dataset(f'anirudhb11/{dataset_name}')

# {"input": "5,9/11,7/2,1/9,3/3,8/7,2/8,4/6,11/5,6/4,0-5,1", "output": "5,6/6,11/11,7/7,2/2,1", "reversed": "1,2/2,7/7,11/11,6/6,5"}

def convert_to_array(dataset):
    arr_dataset = []
    for indx in range(len(dataset)):
        graph =  dataset[indx]['graph']
        path =  dataset[indx]['path']
        src = dataset[indx]['source']
        dst = dataset[indx]['destination']
        nodes_in_path = path.split(',')
        
        
        data_point_dict = {}
        data_point_dict['question'] = ''
        data_point_dict['answer'] = ''
        data_point_dict['steps'] = []
        
        edges = graph.split('|')
        for edge in edges:
            src_vertex, dst_vertex = edge.split(',')
            data_point_dict['question'] = data_point_dict['question'] + f'{src_vertex} is connected to {dst_vertex}. '
        data_point_dict['question'] = data_point_dict['question'] + f'What is the path from {src} to {dst}?'
        
        data_point_dict['answer'] = path
        
        for i in range(1, len(nodes_in_path)):
            data_point_dict['steps'].append(f'{nodes_in_path[i - 1]} is connected to {nodes_in_path[i]}.')
        arr_dataset.append(data_point_dict)

    return arr_dataset
```

```python
train_ds = convert_to_array(dataset['train'])
```

```python
import numpy.random as rand

perm = rand.permutation(len(train_ds))
```

```python
val_ds = []
train_ds_new = []

for indx in perm[:500]:
    print(indx)
    val_ds.append(train_ds[indx])

for indx in perm[500:]:
    train_ds_new.append(train_ds[indx])
    
```

```python
test_ds = convert_to_array(dataset['test'])
```

```python
import json
json.dump(train_ds_new, open(f'../data/{dataset_name}_train.json', 'w'))
```
